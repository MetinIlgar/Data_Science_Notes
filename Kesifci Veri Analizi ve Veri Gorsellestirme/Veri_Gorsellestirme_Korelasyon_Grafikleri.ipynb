## Scatterplot

```python
# Saçılım grafiği
# sayıssal değişkenler arasındaki ilişkiyi gösterir.
```

**Veri Seti Hikayesi**
* total_bill: yemeğin toplam fiyatı (bahşiş ve verigler dahil)
* tip: bahşiş
* sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)
* smoker: grupta sigara içen var mı? (0=no, 1=yes)
* day: gün (3=Thur, 4=Fri, 5=Sat, 6=Sun)
* time: ne zaman? (0=day, 1=night)
* size: grupta kaç kişi var?

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data =df);
# Ödenen hesap arttıkça, bahşiş miktarı da artıyor
```

### Çaprazlamalar

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", style = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day", style = "day", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day", style = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", size = "size", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", size = "size", hue = "size", data = df);
```

## Doğrusal İlişkinin Gösterilmesi

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
df = tips.copy()
df.head()

sns.lmplot(x = "total_bill", y = "tip", data = df);
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", data = df);
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "time", data = df);
```

```python
 sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "time", row = "sex", data = df);
```

## Heatmap

### Veri Seti Hikayesi

```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["passengers"].describe()
```

```python
df = df.pivot("month", "year", "passengers")
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df, annot = True, fmt = "d");
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = .3);
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = .3, cbar = False);
```

## Çizgi Grafiği

### Veri Seti Hikayesi

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal")["timepoint"].count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

### Çizgi Grafiği ve Çaprazlamalar

```python
sns.lineplot(x = "timepoint", y = "signal", data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", style = "event", data = df);
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue = "event", 
             style = "event", 
             markers = True,
             dashes = False,
             data = df); 
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue = "region", 
             style = "event", 
             data = df); 
```
